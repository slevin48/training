```python
from urllib.request import urlopen
import json

def get_forecast():
    url="http://samples.openweathermap.org/data/2.5/forecast?q=Muenchen,DE&appid=b1b15e88fa797225412429c1c50c122a1"
    html = urlopen(url).read()
    json_data = json.loads(html)
    time = [l['dt'] for l in json_data['list']]
    temp = [l['main']['temp'] for l in json_data['list']]
    return time,temp
```

```python
url="http://samples.openweathermap.org/data/2.5/forecast?q=Muenchen,DE&appid=b1b15e88fa797225412429c1c50c122a1"
html = urlopen(url).read()
json_data = json.loads(html)
```

```python
f = open('data.txt','w')
f.write("Time,Temperature\n")
for i in range(len(time)):
    f.write(str(datetime.datetime.fromtimestamp(time[i])))
    f.write(",")
    f.write(str(temp[i]))
    f.write("\n")
f.close()
```
